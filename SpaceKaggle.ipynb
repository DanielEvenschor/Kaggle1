## Problem statement

A unusual space-time anomaly came into contact with Spaceship Titanic, an interstellar passenger spacecraft, in the year of 2912. Many passengers were accidentally transported into alternative dimension as a result of this accident. The crew is experiencing difficulties identifying those who had been transported because the accident had no physical damage on the ship.

The governments of the different Home Planets are currently looking for new ways to identify their people, and they have hired Team 12 as a specialized team to work with the search and rescue operations to find the missing passengers.

The predictive model enables passenger detection that may have been impacted by the space-time anomaly and traveled into an alternate dimension. Governments of respective Home Planets may leverage these insights to the rescue their people. Enterprises can make use of this insights to strengthen security measures as well.  

## Assumptions

- Assumption 1: If two or more passengers have the same last name, they are from the same family.
- Assumption 2: If two or more passengers have the same last name, they are from the same home planet.
- Assumption 3: If two or more passengers are family members, they travel in the same group.

## Features Selection / Engineering

- Age_group: This feature involves dividing the age of passengers into different age groups to make more accurate predictions. The age groups range from 0 to 5, 5-10 and then in steps of 10 years onwards. By dividing the age into age groups, the model can better capture the characteristics and behavior of different age groups and make predictions accordingly.

- Expenditure: This feature involves summing up all the expense variables to create a total expense column. This can help in identifying passengers who spent more than others and how much they spent on different services during their journey.

- No_spending: This feature indicates if the total spending of a passenger is zero. It can be used to identify passengers who did not avail any services during their journey.

- Group and Group_size: This feature involves extracting the group and group size information from passenger data. This information can be used to fill missing values and impute data later on in the analysis.

- Solotravel: This feature indicates if a passenger is travelling alone or not. It can help in identifying the behavior and characteristics of solo travelers and how they differ from group travelers.

- Cabin_side: This feature involves splitting the cabin numbers into groups and distributing them into different regions. This can help in identifying which passengers were accommodated in which part of the ship and how it affected their experience.

- Surname: This feature involves extracting the surname of passengers to identify which people belong to the same group. This can be useful in identifying families and groups of passengers travelling together.

- Family_size: This feature involves extracting the family size information from passenger data. This information can be used to fill missing values and impute data later on in the analysis. It can also help in identifying the behavior and characteristics of passengers travelling with their families.

## Training Method(s)

Gradient Booster Classifier: The Model enables an additive model that allows for optimization on on arbitrary differentiable loss functions. It is important to note that the model's features are always randomly fluctuate at each split (Friedman, 2001). This was the model with the best results for prediction and accuracy. 

Random Forest classifier: Estimates a number of decision tree classifiers on several subsamples of the dataset. The method uses averages to improve accuracy and control over-fitting models (Scikit-Learn, n.d). This model was tested, however not used for the final solution. 

Decision Tree Classifier: Training method that employs data sets to predict outputs based on the highest probabilities (Hastie, Friedman & Tisbshirani, 2017). The model had a lower accuracy, so it was not used for the final solution. 

Logistic Regression: Linear model for classification, the model uses a single trial to predict possible outcomes through logistic functions (Galarnyk, 2017). It is important to mention that the model employs random number generators to fit features into the model. Slightly different results may occur when running the model, however, smaller parameters may limit this outcomes. 

AdaBoost Classifier:Any machine learning algorithm's performance can be improved with AdaBoost. When teaching weak learners, it works best. These are models whose categorization accuracy is slightly better than random chance. Decision trees with one level are the most suitable and, as a result, the most frequently employed algorithm with AdaBoost(Brownlee, 2020).

Extra Tree Classifier: ExtraTrees is an ensemble machine learning approach that, like Random Forests, trains a large number of decision trees and then combines their individual outputs to produce a forecast.To guarantee that the decision trees are sufficiently varied, Random Forest employs bagging to choose various versions of the training data. Extra Trees, on the other hand, trains decision trees using the whole dataset. As a result, it RANDOMLY SELECTS the values at which to split a feature and build child nodes in order to guarantee that there are enough variances between individual decision trees. With a Random Forest, however, we employ an algorithm to greedily explore and choose the value at which to split a feature(Thankachan, 2022).

### Feature Engineering

Age

```python
# according to the bins that was determined earlier 
# Create a new column 'Age_group' and initialize all values to NaN
train['Age_group'] = np.nan

# Boolean indexing to update 'Age_group'  based on age
train.loc[train['Age'] <= 5, 'Age_group'] = 'Age_0-5'
train.loc[(train['Age'] > 5) & (train['Age'] < 10), 'Age_group'] = 'Age_6-9'
train.loc[(train['Age'] >= 10) & (train['Age'] <= 18), 'Age_group'] = 'Age_10-18'
train.loc[(train['Age'] > 18) & (train['Age'] <= 25), 'Age_group'] = 'Age_19-25'
train.loc[(train['Age'] > 25) & (train['Age'] <= 50), 'Age_group'] = 'Age_26-50'
train.loc[train['Age'] > 50, 'Age_group'] = 'Age_51+'

# same for test
test['Age_group'] = np.nan
test.loc[test['Age'] <= 5, 'Age_group'] = 'Age_0-5'
test.loc[(test['Age'] > 5) & (test['Age'] < 10), 'Age_group'] = 'Age_6-9'
test.loc[(test['Age'] >= 10) & (test['Age'] <= 18), 'Age_group'] = 'Age_10-18'
test.loc[(test['Age'] > 18) & (test['Age'] <= 25), 'Age_group'] = 'Age_19-25'
test.loc[(test['Age'] > 25) & (test['Age'] <= 50), 'Age_group'] = 'Age_26-50'
test.loc[test['Age'] > 50, 'Age_group'] = 'Age_51+'
```

All expenditures

```python
# Creaete new collumn Expenditure and No_spending
# Expenditure should be the total of all the possible expenses on board
# No_spending should be 1/0 true and false if the total expenditure is 0 or not 
# All expenditure features
exp_feats=['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']

train['Expenditure'] = train[exp_feats].sum(axis=1) # sum all expenditure features for total expenditure

train['No_spending'] = (train['Expenditure'] == 0).astype(int) # if total is 0 then true for no_spending
# same for test set
test['Expenditure']=test[exp_feats].sum(axis=1)  
test['No_spending']=(test['Expenditure']==0).astype(int)

```

Grop Travellers, Solo Travellers

```python
# IDENTIFY THE GROUPS FROM PASSENGER ID
# PassengerId : xxxx_yy ; xxxx indicates a group the passenger is travelling with
# yy is their number within the group

# split the passenger ID by seperator '_'
train['Group'] = train['PassengerId'].str.split('_').str[0].astype(int) # select only the xxxx and store it in "Group"
test['Group'] = test['PassengerId'].str.split('_').str[0].astype(int)

# groups are useless. Decision to make a collumn called solo for true if they are not in a group

combined_group = pd.concat([train['Group'], test['Group']]) # combine group for ease of use, then run value_counts on group
group_counts = combined_group.value_counts()
train['solotravel'] = (train['Group'].map(group_counts) == 1).astype(int) # use map according to the value counts, check if its ==1 and return boolean 
test['solotravel'] = (test['Group'].map(group_counts) == 1).astype(int)
```

Cabin

```python
# Cabin

# Replace NaN's with clear fakes so i can split
train['Cabin'].fillna('Z/9999/Z', inplace=True)
test['Cabin'].fillna('Z/9999/Z', inplace=True)

# Split the cabin collumn to get 3 new collumns for both train and test - 
# Cabin : deck/num/side, where side can be either P for Port or S for Starboard

train['Cabin_deck'] = train['Cabin'].apply(lambda x: x.split('/')[0]) # [0] for the First Letter 
train['Cabin_number'] = train['Cabin'].apply(lambda x: x.split('/')[1]).astype(int) # [1] for the Number
train['Cabin_side'] = train['Cabin'].apply(lambda x: x.split('/')[2]) #[2] for cabin side (last value in the split)
# same for test
test['Cabin_deck'] = test['Cabin'].apply(lambda x: x.split('/')[0])
test['Cabin_number'] = test['Cabin'].apply(lambda x: x.split('/')[1]).astype(int)
test['Cabin_side'] = test['Cabin'].apply(lambda x: x.split('/')[2])

# add NaN values back using the clear fakes for both 
train.loc[train['Cabin_deck']=='Z', 'Cabin_deck']=np.nan
train.loc[train['Cabin_number']==9999, 'Cabin_number']=np.nan
train.loc[train['Cabin_side']=='Z', 'Cabin_side']=np.nan
test.loc[test['Cabin_deck']=='Z', 'Cabin_deck']=np.nan
test.loc[test['Cabin_number']==9999, 'Cabin_number']=np.nan
test.loc[test['Cabin_side']=='Z', 'Cabin_side']=np.nan

# Drop Cabin since we have split the content and now have 3 collumns instead of 1 
train.drop('Cabin', axis=1, inplace=True)
test.drop('Cabin', axis=1, inplace=True)
```

Name

```python
# Final qualitative is name 
# Name seems useless from the beginning i wanted to drop it 
# Extracting surname has proven to be useful so ... same approach as with cabin

# Replace NaN with fakes 
train['Name'].fillna('Unknown Unknown', inplace=True)
test['Name'].fillna('Unknown Unknown', inplace=True)

# Create collumn Surname split and select surname [-1], assuming that the names are not "Surname Name" in some cases .. wont check 
train['Surname']=train['Name'].str.split().str[-1]
test['Surname']=test['Name'].str.split().str[-1]

# groupby surname and then count the amount of passenger ID's 
# Create new collumn Family size
train['Family_size'] = train.groupby('Surname')['PassengerId'].transform(len)
test['Family_size'] = test.groupby('Surname')['PassengerId'].transform(len)

# Put missing values back using the fakes
train.loc[train['Surname']=='Unknown','Surname']=np.nan
train.loc[train['Family_size']>100,'Family_size']=np.nan # there shouldnt be a family bigger than 100 members 
test.loc[test['Surname']=='Unknown','Surname']=np.nan
test.loc[test['Family_size']>=94,'Family_size']=np.nan # set the group of 94 nulls to nan instead of huge family of all the nulls

# Drop name collumn because we extracted everything we need
train.drop('Name', axis=1, inplace=True)
test.drop('Name', axis=1, inplace=True)
```

```python
# Filling in the HomePlante missing values 

# get the count for homeplanet using .size and then unstack the homeplanet and group into a df, fill missing values with 0 
grouphomeplanet=data.groupby(['Group','HomePlanet'])['HomePlanet'].size().unstack().fillna(0)
grouphomeplanet # INSIGHT: Passengers from the same homeplanet are in the same group 

# NaN before
HP_bef=data['HomePlanet'].isna().sum()

# everyone in the same group comes from the same home planet
# So we can fill the missing HomePlanet values according to the group
# If the value in homeplanet is NaN, look into the unstack
# select all of the rows that are NaN in Home planet but have are in a group, match with the index of grouphomeplanet that contains all groups
index=data[data['HomePlanet'].isna()][(data[data['HomePlanet'].isna()]['Group']).isin(grouphomeplanet.index)].index

# Fill in the missing values based on the index variable that only contains the rows that match the condition
groups_to_homeplanet = dict(grouphomeplanet.idxmax(axis=1))
data.loc[index, 'HomePlanet'] = data.loc[index, 'Group'].map(groups_to_homeplanet)

# missing values left
print('#HomePlanet missing values before:',HP_bef)
print('#HomePlanet missing values after:',data['HomePlanet'].isna().sum())
```

```python
#The cabin_number and group_number share a linear relationship on a deck by deck basis. 
# Predict the missing cabin numbers using linear regression cabin number.

# NaN before
CN_bef=data['Cabin_number'].isna().sum()

#  linear relationship on a deck by deck basis
for deck in ['A', 'B', 'C', 'D', 'E', 'F', 'G']:
    X_CN=data.loc[~(data['Cabin_number'].isna()) & (data['Cabin_deck']==deck),'Group']
    y_CN=data.loc[~(data['Cabin_number'].isna()) & (data['Cabin_deck']==deck),'Cabin_number']
    X_test_CN=data.loc[(data['Cabin_number'].isna()) & (data['Cabin_deck']==deck),'Group']

    # apply Linear regression
    model_CN=LinearRegression()
    model_CN.fit(X_CN.values.reshape(-1, 1), y_CN)
    preds_CN=model_CN.predict(X_test_CN.values.reshape(-1, 1))
    
    # Fill missing values with predictions
    data.loc[(data['Cabin_number'].isna()) & (data['Cabin_deck']==deck),'Cabin_number']=preds_CN.astype(int)

# No values left
print('#Cabin_number missing values before:',CN_bef)
print('#Cabin_number missing values after:',data['Cabin_number'].isna().sum())
```
